```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytaj dane z pliku CSV
file_path = 'temperatura.csv'  # Upewnij się, że ścieżka jest poprawna
data = pd.read_csv(file_path)

# Konwersja kolumny daty do formatu datetime
data['category'] = pd.to_datetime(data['category'], format='%d.%m.%Y')

# Filtracja danych dla stycznia 2012
january_2012_data = data[(data['category'].dt.month == 1) & (data['category'].dt.year == 2012)]

# Ustawienie dat jako indeks
january_2012_data.set_index('category', inplace=True)

# Tworzenie wykresu
plt.figure(figsize=(10, 5))
plt.plot(january_2012_data.index, january_2012_data['minimum'], label='Minimum Temperature', color='blue', marker='o')
plt.plot(january_2012_data.index, january_2012_data['maximum'], label='Maximum Temperature', color='red', marker='o')

# Dodanie tytułu i etykiet
plt.title('Temperature Changes in January 2012')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

# Wyświetlenie wykresu
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytaj dane z pliku CSV
file_path = 'temperatura.csv'  # Upewnij się, że ścieżka jest poprawna
data = pd.read_csv(file_path)

# Konwersja kolumny daty do formatu datetime
data['category'] = pd.to_datetime(data['category'], format='%d.%m.%Y')

# Filtracja danych dla stycznia 2012 i 2024
january_2012_data = data[(data['category'].dt.month == 1) & (data['category'].dt.year == 2012)]
january_2024_data = data[(data['category'].dt.month == 1) & (data['category'].dt.year == 2024)]

# Sprawdzenie, czy mamy dane do porównania
if not january_2012_data.empty and not january_2024_data.empty:
    # Przygotowanie danych do wykresu
    dates = january_2012_data['category'].dt.day
    min_2012 = january_2012_data['minimum'].values
    max_2012 = january_2012_data['maximum'].values
    min_2024 = january_2024_data['minimum'].values
    max_2024 = january_2024_data['maximum'].values

    # Tworzenie wykresu
    plt.figure(figsize=(10, 5))

    # Wykres dla stycznia 2012
    plt.plot(dates, min_2012, label='Minimum Temperature 2012', color='blue', marker='o')
    plt.plot(dates, max_2012, label='Maximum Temperature 2012', color='red', marker='o')

    # Wykres dla stycznia 2024
    plt.plot(dates, min_2024, label='Minimum Temperature 2024', color='cyan', marker='o')
    plt.plot(dates, max_2024, label='Maximum Temperature 2024', color='orange', marker='o')

    # Dodanie tytułu i etykiet
    plt.title('Temperature Comparison: January 2012 vs January 2024')
    plt.xlabel('Day of January')
    plt.ylabel('Temperature (°C)')
    plt.xticks(dates)  # Ustawienie etykiet dni
    plt.legend()
    plt.grid()

    # Wyświetlenie wykresu
    plt.tight_layout()
    plt.show()
else:
    print("Brak danych dla jednego z lat.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Wczytaj dane z pliku CSV
file_path = 'temperatura.csv'  # Upewnij się, że ścieżka jest poprawna
data = pd.read_csv(file_path)

# Konwersja kolumny daty do formatu datetime
data['category'] = pd.to_datetime(data['category'], format='%d.%m.%Y')

# Filtracja danych dla stycznia (wszystkie lata)
january_data = data[data['category'].dt.month == 1]

# Ekstrakcja roku i obliczenie średnich temperatur
january_data['year'] = january_data['category'].dt.year
average_temperatures = january_data.groupby('year').agg({'minimum': 'mean', 'maximum': 'mean'}).reset_index()

# Przygotowanie danych do regresji
X = average_temperatures[['year']]
y_min = average_temperatures['minimum']
y_max = average_temperatures['maximum']

# Tworzenie modelu regresji
model_min = LinearRegression()
model_max = LinearRegression()

# Trenowanie modelu
model_min.fit(X, y_min)
model_max.fit(X, y_max)

# Prognoza na przyszłość (np. do roku 2030)
future_years = np.arange(2024, 2031).reshape(-1, 1)
predicted_min = model_min.predict(future_years)
predicted_max = model_max.predict(future_years)

# Połączenie z danymi historycznymi do wykresu
combined_years = np.concatenate((average_temperatures['year'].values, future_years.flatten()))
combined_min = np.concatenate((y_min.values, predicted_min))
combined_max = np.concatenate((y_max.values, predicted_max))

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
plt.plot(average_temperatures['year'], y_min, label='Average Minimum Temperature', color='blue', marker='o')
plt.plot(average_temperatures['year'], y_max, label='Average Maximum Temperature', color='red', marker='o')
plt.plot(future_years, predicted_min, label='Predicted Minimum Temperature', color='lightblue', linestyle='--')
plt.plot(future_years, predicted_max, label='Predicted Maximum Temperature', color='lightcoral', linestyle='--')

# Dodanie tytułu i etykiet
plt.title('January Temperature Trends and Predictions')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.xticks(np.arange(2012, 2031, 1))
plt.legend()
plt.grid()

# Wyświetlenie wykresu
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Wczytaj dane z pliku CSV
file_path = 'temperatura.csv'  # Upewnij się, że ścieżka jest poprawna
data = pd.read_csv(file_path)

# Konwersja kolumny daty do formatu datetime
data['category'] = pd.to_datetime(data['category'], format='%d.%m.%Y')

# Ekstrakcja roku i miesiąca
data['year'] = data['category'].dt.year
data['month'] = data['category'].dt.month

# Filtracja danych dla sezonu letniego (kwiecień do października)
summer_data = data[data['month'].isin([4, 5, 6, 7, 8, 9, 10])]

# Obliczanie średnich temperatur dla każdego roku w sezonie letnim
average_summer_temperatures = summer_data.groupby('year').agg({'minimum': 'mean', 'maximum': 'mean'}).reset_index()

# Przygotowanie danych do regresji
X = average_summer_temperatures[['year']]
y_min = average_summer_temperatures['minimum']
y_max = average_summer_temperatures['maximum']

# Tworzenie modelu regresji
model_min = LinearRegression()
model_max = LinearRegression()

# Trenowanie modelu
model_min.fit(X, y_min)
model_max.fit(X, y_max)

# Prognoza na przyszłość (np. do roku 2030)
future_years = np.arange(2024, 2031).reshape(-1, 1)
predicted_min = model_min.predict(future_years)
predicted_max = model_max.predict(future_years)

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
plt.plot(average_summer_temperatures['year'], y_min, label='Average Minimum Temperature (Apr-Oct)', color='blue', marker='o')
plt.plot(average_summer_temperatures['year'], y_max, label='Average Maximum Temperature (Apr-Oct)', color='red', marker='o')
plt.plot(future_years, predicted_min, label='Predicted Minimum Temperature', color='lightblue', linestyle='--')
plt.plot(future_years, predicted_max, label='Predicted Maximum Temperature', color='lightcoral', linestyle='--')

# Dodanie tytułu i etykiet
plt.title('Average Temperature Trends (April - October) and Predictions')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.xticks(np.arange(int(average_summer_temperatures['year'].min()), 2031, 1))
plt.legend()
plt.grid()

# Wyświetlenie wykresu
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Wczytaj dane z pliku CSV
file_path = 'temperatura.csv'  # Upewnij się, że ścieżka jest poprawna
data = pd.read_csv(file_path)

# Konwersja kolumny daty do formatu datetime
data['category'] = pd.to_datetime(data['category'], format='%d.%m.%Y')

# Ekstrakcja roku i miesiąca
data['year'] = data['category'].dt.year
data['month'] = data['category'].dt.month

# Filtracja danych dla sezonu zimowego (grudzień, styczeń, luty)
winter_data = data[data['month'].isin([12, 1, 2])]

# Obliczanie średnich temperatur dla każdego roku w sezonie zimowym
average_winter_temperatures = winter_data.groupby('year').agg({'minimum': 'mean', 'maximum': 'mean'}).reset_index()

# Przygotowanie danych do regresji
X = average_winter_temperatures[['year']]
y_min = average_winter_temperatures['minimum']
y_max = average_winter_temperatures['maximum']

# Tworzenie modelu regresji
model_min = LinearRegression()
model_max = LinearRegression()

# Trenowanie modelu
model_min.fit(X, y_min)
model_max.fit(X, y_max)

# Prognoza na przyszłość (np. do roku 2030)
future_years = np.arange(2024, 2031).reshape(-1, 1)
predicted_min = model_min.predict(future_years)
predicted_max = model_max.predict(future_years)

# Tworzenie wykresu
plt.figure(figsize=(12, 6))
plt.plot(average_winter_temperatures['year'], y_min, label='Average Minimum Temperature (Winter)', color='blue', marker='o')
plt.plot(average_winter_temperatures['year'], y_max, label='Average Maximum Temperature (Winter)', color='red', marker='o')
plt.plot(future_years, predicted_min, label='Predicted Minimum Temperature', color='lightblue', linestyle='--')
plt.plot(future_years, predicted_max, label='Predicted Maximum Temperature', color='lightcoral', linestyle='--')

# Dodanie tytułu i etykiet
plt.title('Average Winter Temperature Trends and Predictions')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.xticks(np.arange(int(average_winter_temperatures['year'].min()), 2031, 1))
plt.legend()
plt.grid()

# Wyświetlenie wykresu
plt.tight_layout()
plt.show()
```
